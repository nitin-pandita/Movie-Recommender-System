```python
# we can create a function that will convert the dictionary to a list
import ast # this function will help us to convert the indexes to string otherwise it will give us error
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): #function literal_eval help us to do so
       L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movie['cast'] = movie['cast'].apply(convert3)
```

```python
movie.head()
```

# Getting the Director Name from the dataset

```python
# we will create a function for that
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

# converting the overiew to a string using the lambda function

```python
movie['overview'][0]
```

```python
movie['overview'] = movie['overview'].apply(lambda x:str(x).split())
```

```python
movie.head()
```

# replacing the space with no space as it may cause confusion in finding the particular recomended movie

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tag'] = movie['overview'] + movie['keywords'] + movie['genres'] + movie['cast'] + movie['crew']
```

```python
new_Df = movie[['movie_id', 'title','tag']]
new_Df
```

# Converting the list to string for "Tags"

```python
new_Df['tag'] = new_Df['tag'].apply(lambda x:" ".join(x))
```

# Creating the vectors

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_Df['tag']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# helper function
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
        
```

```python
new_Df['tag'] = new_Df['tag'].apply(stem)
```

```python
new_Df
```

# findig the similarity in the movies 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

# i have to make then the arrangement in descending order

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] # we will be using enumerate for getting the tuple
```

```python
# remcomendation function
def recommendation(movie):
    movie_index = new_Df[new_Df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_Df.iloc[i[0]].title)
```

```python
recommendation('Batman Begins')
```
